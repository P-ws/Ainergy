```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings
import datetime
warnings.filterwarnings('ignore')
plt.rc('font', family='Malgun Gothic') 
```

```python
# datetime을 index로 옮겨주는 함수
def datetime_revise(df):
    
    # python datetime에서는 시간을 0~23시까지밖에 인식을 못하기 때문에 1을 빼준다.
    df.index = pd.to_datetime(df['datetime']) + pd.to_timedelta(df['hour'], unit='hours') - pd.to_timedelta(1, unit='hour')
    df.drop(['year', 'month', 'day', 'weekday', 'hour', 'datetime'], axis=1, inplace=True)
    return df
```

```python
# 시계열 분석을 위해 target값만을 뽑음
def make_df(df, colname):
    df = total.query(f'supplys == "{colname}"')
    return df['target']
```

```python
# upsamplingm downsampling해줄 수 있도록 하는 함수
def set_freq(df, freq='m'):
    return df.resample(freq).sum()
```

```python
total = pd.read_csv('data/no_outlier_data.csv')
total = datetime_revise(total)
total
```

```python
# 공급사 이름
supply_name_list = [n for n in total['supplys'].unique().tolist()]
supply_name_list
```

## 시계열 데이터 탐색

```python
# acf, pacf graph
import statsmodels.graphics.tsaplots as sgt
```

```python
import statsmodels.tsa.stattools as sts

# 한 번에 분석하기 용이하도록 각 공급사별로 일별 공급 데이터만 담긴 series를 따로 저장
colname = total['supplys'].unique().tolist()
```

### 1. 일 단위 탐색

```python
day_time_series = [set_freq(make_df(total, col), freq='d') for col in colname]
day_time_series[0]
```

```python
# 일별 시계열 데이터를 도식화 (3개월만 표시)
plt.subplots(figsize=(20, 15))

for idx, t_df in enumerate(day_time_series):
    ax = plt.subplot(4, 2, idx+1)
    t_df.plot(xlim=[datetime.date(2013, 1, 1), datetime.date(2013, 3, 31)])
```

#### 잔차진단
- 잔차는 Y - Trend - Seasonal에서 만들어지는데, 역으로 생각하면 만약 잔차에 trend나 seasonal 특성이 남아있으면, 완벽하게 trend, seasonality 부분이 추출되지 않았음을 의미
- 잔차는 백색잡음(white noise) 특성을 가져야 함
- 백색잡음 특징 : 평균 0, 분산 일정, 정규분포 따라야 함, 자기상관도함수 값 0이어야
- 출처 : https://dsbook.tistory.com/265?category=841791

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# 분해하보자
result = seasonal_decompose(day_time_series[0], model='additive')
result.plot()
plt.show()

# 딱 봐도 잔차에서 패턴이 있는 것을 볼 수 있다.
```

```python
# 잔차 추출
residual = result.resid
# residual.isna().sum()
residual.fillna(0, inplace=True)
```

```python
# 1. 잔차의 정규분포 test
from scipy.stats import shapiro

# p-value가 0.05보다 작으므로 정규분포가 아니라는 검정 
shapiro(residual)
```

```python
# 2. 등분산성 test(귀무가설 test)
# 각 시차에 따라 분산이 일정하게 유지되는지 test

```

```python
result.observed.shape
```

```python
# 3. 자기상관 test(Djung-box)
# 귀무가설 : 자기상관성이 없다
# 대립가설 : 자기상관성이 있다.
#모든 lag에 대해 자기상관성이 있는 것을 확인할 수 있다

from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
acorr_ljungbox(residual, return_df=True)
```

#### ADF TEST를 통한 단위근검정
- p-value가 유의수준보다 작으면 귀무가설(단위근이 있다) 기각
- 주의할 점 : 단위근이 있으면 비정상이라고 할 수 있지만, 단위근이 없다고 해서 정상 시계열은 아니다
- 추가적인 검증이 필요함

```python
# 유의수준 설정
alpha = 0.05

print('')
for idx, t_df in enumerate(day_time_series):

    p_value = sts.adfuller(t_df)[1]
    
    # 귀무가설 기각
    if p_value < alpha:
        print(f'{supply_name_list[idx]}는 p-value({round(p_value, 3)})가 유의수준 {alpha}보다 낮으므로 단위근이 존재하지 않습니다')
    else:
        print(f'(비정상 시계열) {supply_name_list[idx]}는 p-value({round(p_value, 3)})가 유의수준 {alpha}보다 높으므로 단위근이 존재합니다')
```

#### KPSS TEST를 통한 정상성 검정
- ADF TEST와는 반대로 귀무가설이 정상성, 대립가설이 비정상성이다.
- p-value가 유의수준보다 작으면, 비정상
- ADF에서 발생하는 1종 오류와 2종 오류는 빈번하기 때문에 같이 해주는 게 좋음

```python
alpha = 0.05

print('')
for idx, t_df in enumerate(day_time_series):
    
    #p-value
    p_value = sts.kpss(t_df, nlags='auto')[1]
    
    # p-value가 유의수준보다 낮을 경우 => 정상 시계열이란 귀무가설 기각
    if result < alpha:
        print(f'{supply_name_list[idx]}는 p-value({round(p_value, 3)})가 유의수준 {alpha}보다 낮으므로 비정상 시계열입니다')
    else:
        print(f'{supply_name_list[idx]}는 p-value({round(p_value, 3)})가 유의수준 {alpha}보다 높으므로 정상 시계열입니다.')
```

#### ACF / PACF
- 일 단위의 시계열

```python
import matplotlib.gridspec as gridspec

# Grid하기 위한 작업
fig = plt.figure(figsize=(30, 20))
outer = gridspec.GridSpec(4, 2, wspace=0.2, hspace=0.2)

for idx, t_df in enumerate(day_time_series):
    inner = gridspec.GridSpecFromSubplotSpec(2, 1,
                    subplot_spec=outer[idx], wspace=0.3, hspace=0.3)
    
    # Grid내부에 acf, pacf를 그릴 subplot 생성
    ax1 = plt.Subplot(fig, inner[0])
    ax2 = plt.Subplot(fig, inner[1])
    
    # acf draw
    sgt.plot_acf(t_df, lags = 20, zero = False, ax=ax1)
    ax1.set_title("ACF")
    
    # pacf draw
    sgt.plot_pacf(t_df, lags = 20, zero = False, method = ('ols'), ax=ax2)
    ax2.set_title("PACF")
    
    fig.add_subplot(ax1)
    fig.add_subplot(ax2)

fig.show()
```

#### ADF TEST를 통한 단위근검정
 - 다 비정상 시계열

```python
# 유의수준 설정
alpha = 0.05

print('')
for idx, t_df in enumerate(month_time_series):

    result = sts.adfuller(t_df)[1]
    
    if result < alpha:
        print(f'{supply_name_list[idx]}는 p-value({round(result, 3)})가 유의수준 {alpha}보다 낮으므로 단위근이 존재하지 않습니다')
    else:
        print(f'(비정상 시계열) {supply_name_list[idx]}는 p-value({round(result, 3)})가 유의수준 {alpha}보다 높으므로 단위근이 존재합니다')
```

#### KPSS TEST를 통한 정상성 검정
- ADF TEST와는 반대로 귀무가설이 정상성, 대립가설이 비정상성이다.
- ADF에서 발생하는 1종 오류와 2종 오류는 빈번하기 때문에 같이 해주는 게 좋음

```python
alpha = 0.05

print('')
for idx, t_df in enumerate(month_time_series):

    result = sts.kpss(t_df)[1]
    # p-value가 유의수준보다 낮을 경우 => 기각(비정상 시계열)
    if result < alpha:
        print(f'{supply_name_list[idx]}는 p-value({round(result, 3)})가 유의수준 {alpha}보다 낮으므로 비정상 시계열입니다')
    else:
        print(f'{supply_name_list[idx]}는 p-value({round(result, 3)})가 유의수준 {alpha}보다 높으므로 정상 시계열입니다.')
```
