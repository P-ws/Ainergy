```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
plt.rc('font', family='Malgun Gothic') 
```

```python
# datetime을 index로 옮겨주는 함수
def datetime_revise(df):
    
    # python datetime에서는 시간을 0~23시까지밖에 인식을 못하기 때문에 1을 빼준다.
    df.index = pd.to_datetime(df['datetime']) + pd.to_timedelta(df['hour'], unit='hours') - pd.to_timedelta(1, unit='hour')
    df.drop(['year', 'month', 'day', 'weekday', 'hour', 'datetime'], axis=1, inplace=True)
    return df
```

```python
# 시계열 분석을 위해 target값만을 뽑음
def make_df(df, colname):
    df = total.query(f'supplys == "{colname}"')
    return df['target']
```

```python
# upsamplingm downsampling해줄 수 있도록 하는 함수
def set_freq(df, freq='m'):
    return df.resample(freq).sum()
```

```python
total = pd.read_csv('data/no_outlier_data.csv')
total = datetime_revise(total)
total
```

```python
# 공급사 이름
supply_name_list = [n for n in total['supplys'].unique().tolist()]
supply_name_list
```

## 시계열 데이터 탐색

```python
import statsmodels.tsa.stattools as sts

# 한 번에 분석하기 용이하도록 각 공급사별로 일별 공급 데이터만 담긴 series를 따로 저장
colname = total['supplys'].unique().tolist()
```

### 1. 일 단위 탐색

```python
day_time_series = [set_freq(make_df(total, col), freq='d') for col in colname]
day_time_series[0]
```

```python
# 일별 시계열 데이터를 도식화
plt.subplots(figsize=(20, 15))

for idx, t_df in enumerate(day_time_series):
    ax = plt.subplot(4, 2, idx+1)
    t_df.plot()
```

#### ADF TEST를 통한 단위근검정
- p-value가 유의수준보다 작으면 귀무가설(단위근이 있다) 기각
- 주의할 점 : 단위근이 있으면 비정상이라고 할 수 있지만, 단위근이 없다고 해서 정상 시계열은 아니다
- 추가적인 검증이 필요함

```python
# 유의수준 설정
alpha = 0.05

print('')
for idx, t_df in enumerate(day_time_series):

    result = sts.adfuller(t_df)[1]
    
    if result < alpha:
        print(f'{supply_name_list[idx]}는 p-value({round(result, 3)})가 유의수준 {alpha}보다 낮으므로 단위근이 존재하지 않습니다')
    else:
        print(f'(비정상 시계열) {supply_name_list[idx]}는 p-value({round(result, 3)})가 유의수준 {alpha}보다 높으므로 단위근이 존재합니다')
```

#### ACF / PACF
- 일 단위의 시계열

```python
import matplotlib.gridspec as gridspec

# Grid하기 위한 작업
fig = plt.figure(figsize=(30, 20))
outer = gridspec.GridSpec(4, 2, wspace=0.2, hspace=0.2)

for idx, t_df in enumerate(day_time_series):
    inner = gridspec.GridSpecFromSubplotSpec(2, 1,
                    subplot_spec=outer[idx], wspace=0.3, hspace=0.3)
    
    # Grid내부에 acf, pacf를 그릴 subplot 생성
    ax1 = plt.Subplot(fig, inner[0])
    ax2 = plt.Subplot(fig, inner[1])
    
    # acf draw
    sgt.plot_acf(t_df, lags = 20, zero = False, ax=ax1)
    ax1.set_title("ACF")
    
    # pacf draw
    sgt.plot_pacf(t_df, lags = 20, zero = False, method = ('ols'), ax=ax2)
    ax2.set_title("PACF")
    
    fig.add_subplot(ax1)
    fig.add_subplot(ax2)

fig.show()
```
